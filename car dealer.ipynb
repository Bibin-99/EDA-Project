# CAR DEALERS

### The dataset represents transactions at a car dealership, recording details about car sales, customers, and dealers. It includes customer information such as name, email, phone number, and city, along with car details like brand, model, year, color, and price. Each transaction is tied to a specific dealer and salesperson, with additional details like payment method and warranty period. The dataset captures the date and time of each transaction, making it suitable for sales trend analysis. It provides a comprehensive view of dealership operations, customer demographics, and sales performance.

# QUESTIONS

### 1.What is the total revenue generated from car sales (Price)?

```python
total_revenue = df['Price'].sum()
print(total_revenue)
```

### 2.What is the average car price?

```python
average = df['Price'].mean()
print(average)
```

### 3. How many cars were sold for each brand (Car_Brand)?

```python
cars_per_brand = df['Car_Brand'].value_counts()
print(cars_per_brand)
```

### 4. What is the distribution of car prices?

```python
sns.histplot(df['Price'], bins=30, kde=True)
```

### 5. Which salesperson generated the most revenue?

```python
revenue_per_salesperson = df.groupby('Salesperson_Name')['Price'].sum().idxmax()
print(revenue_per_salesperson)
```

### 6. How does car price vary by brand (Car_Brand)?

```python
sns.boxplot(data=df, x='Car_Brand', y='Price')
```

### 7. What is the relationship between car year (Car_Year) and price?

```python
sns.scatterplot(data=df, x='Car_Year', y='Price')
```

### 8. Are newer cars (Car_Year) sold for higher prices on average?

```python
avg_price_by_year = df.groupby('Car_Year')['Price'].mean()
print(avg_price_by_year)
```

### 9. How does the warranty period (Warranty_Period) correlate with car price?

```python
sns.boxplot(data=df, x='Warranty_Period', y='Price')
```

### 10. What is the trend of sales (Price) over time (Date)?

```python
df['Date'] = pd.to_datetime(df['Date'])
sales_over_time = df.groupby('Date')['Price'].sum()
sales_over_time.plot()
```

### 11.	Are there any significant outliers in car prices? 

```python
sns.boxplot(data=df, x='Price')
```

### 12.	What is the standard deviation of car prices?

```python
std_dev = df['Price'].std()
print("Standard Deviation:", std_dev)
```

### 13.	Which car model (Car_Model) is the most expensive?

```python
most_expensive_model = df.loc[df['Price'].idxmax(), 'Car_Model']
print(most_expensive_model)
```

### 14.	What is the average price for each car color (Car_Color)?

```python
avg_price_by_color = df.groupby('Car_Color')['Price'].mean()
print(avg_price_by_color)
```

### 15.	How many cars were sold under each payment method (Payment_Method)?

```python
cars_by_payment = df['Payment_Method'].value_counts()
print(cars_by_payment)
```

### 16.	Which city (Customer_City) purchased the most cars?

```python
top_city = df['Customer_City'].value_counts()
print(top_city)
```

### 17.	What is the distribution of warranty periods (Warranty_Period)?

```python
sns.countplot(data=df, x='Warranty_Period')
```

### 18.	Is there a significant difference in prices between Toyota and Audi cars?

```python
toyota_prices = df[df['Car_Brand'] == 'Toyota']['Price']
audi_prices = df[df['Car_Brand'] == 'Audi']['Price']
from scipy.stats import ttest_ind
t_stat, p_val = ttest_ind(toyota_prices, audi_prices)
print(toyota_prices)
print(audi_prices)
```

### 19.	Which customer purchased the most expensive car?

```python
top_customer = df.loc[df['Price'].idxmax(), 'Customer_Name']
print(top_customer)
```

### 20.	What is the median car price for each warranty period (Warranty_Period)?

```python
median_price_by_warranty = df.groupby('Warranty_Period')['Price'].median()
print(median_price_by_warranty)
```

# PIE CHART REPRESENTATION
