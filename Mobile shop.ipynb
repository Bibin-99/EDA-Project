# MOBILE SHOP

### The dataset is a detailed log of transactions at a mobile shop, containing 1,000 records across 15 columns. Key columns include Transaction_ID, Date, Time, Customer_Name, Mobile_Brand, Model, Price, Quantity, Payment_Method, Salesperson_Name, and Store_Location, which capture transactional specifics. Additionally, there are details on warranties (Warranty_Years), extras purchased (Accessories_Purchased), customer feedback scores (Customer_Feedback), and discounts applied (Discount_Applied). For instance, each row represents a unique transaction with information such as the mobile model purchased, payment method used, salesperson involved, and any discounts. Sample entries indicate diversity in customer preferences, store locations, and purchase patterns, making the dataset suitable for sales trend analysis, customer behavior insights, and operational optimization.

# QUESTIONS

### 1.Plot a histogram of a numeric column (e.g., price).

```python
plt.hist(df['Price'], bins=20, color='blue', edgecolor='black')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

### 2.Calculate the mean and median price of mobile phones.

```python
mean_price = np.mean(df['Price'])
median_price = np.median(df['Price'])
print(f"Mean Price: {mean_price}, Median Price: {median_price}")
```

### 3.Display the correlation matrix.

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

# HEAT MAP

```python
sns.heatmap(correlation,annot=True,cmap='coolwarm')
```

### 4.Create a boxplot for a categorical column vs. price (e.g., brand).

```python
sns.boxplot(x='Mobile_Brand', y='Price', data=df)
plt.title("Price Distribution by Brand")
plt.xticks(rotation=45)
plt.show()
```

### 5.Count the number of mobile phones per brand.

```python
brand_counts = df['Mobile_Brand'].value_counts()
print(brand_counts)
```

### 6.Visualize the number of mobile phones per brand using a bar plot.

```python
brand_counts.plot(kind='bar', color='skyblue')
plt.title("Number of Phones per Brand")
plt.xlabel("Mobile_Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### 7.Filter mobile phones priced above a certain threshold (e.g., $500).

```python
high_priced_phones = df[df['Price'] > 500]
print(high_priced_phones)
```

### 8.Group by brand and calculate the average price.

```python
avg_price_per_brand = df.groupby('Mobile_Brand')['Price'].mean()
print(avg_price_per_brand)
```

### 9.Find the most expensive and cheapest mobile phones.

```python
most_expensive = df[df['Price'] == df['Price'].max()]
cheapest = df[df['Price'] == df['Price'].min()]
print("Most Expensive Mobile(s):")
print(most_expensive)
print("\nCheapest Mobile(s):")
print(cheapest)
```

### 10.Plot a line chart of price trends over time (if theres a date column).

```python
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df.sort_values('Date', inplace=True)
    plt.plot(df['Date'], df['Price'], label='Price')
    plt.title("Price Trends Over Time")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.show()
else:
    print("Column 'Date' not found!")
```

### 11.Calculate and visualize the top 5 brands by sales.

```python
if 'Quantity' in df.columns:
    brand_sales = df.groupby('Mobile_Brand')['Quantity'].sum().sort_values(ascending=False).head(5)
    brand_sales.plot(kind='bar', color='green')
    plt.title("Top 5 Brands by Sales")
    plt.xlabel("Mobile_Brand")
    plt.ylabel("Total Sales")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Column 'Quantity' not found!")
```

### 12. Find the variance and standard deviation of prices.

```python
price_variance = np.var(df['Price'])
price_std = np.std(df['Price'])
print(f"Price Variance: {price_variance}, Price Std Dev: {price_std}")
```

### 13.Calculate the percentage distribution of mobile phones across different brands.

```python
brand_percentage = (df['Mobile_Brand'].value_counts(normalize=True) * 100).round(2)
print(brand_percentage)
```

### 14. Salesperson with the highest total revenue

```python
# Calculate total revenue per salesperson
df['Total_Revenue'] = df['Price'] * df['Quantity']
top_salesperson = df.groupby('Salesperson_Name')['Total_Revenue'].sum().sort_values(ascending=False)

# Display and plot the result
print(top_salesperson)
top_salesperson.plot(kind='bar', title="Top Salesperson by Revenue", color='green')
plt.xlabel("Salesperson")
plt.ylabel("Total Revenue")
plt.show()
```

### 15.Total discount applied across different payment methods

```python
# Group by 'Payment_Method' and sum up 'Discount_Applied'
discount_by_payment = df.groupby('Payment_Method')['Discount_Applied'].sum()
# Display and plot the result
print(discount_by_payment)
discount_by_payment.plot(kind='bar', title="Total Discount by Payment Method", color='purple')
plt.xlabel("Payment Method")
plt.ylabel("Total Discount Applied")
plt.show()
```

### 16.Trend of total revenue over time

```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Group by 'Date' and calculate total revenue
revenue_by_date = df.groupby('Date')['Total_Revenue'].sum()

# Plot the revenue trend over time
plt.figure(figsize=(10,6))
plt.plot(revenue_by_date.index, revenue_by_date.values, marker='o', linestyle='-', color='blue')
plt.title("Total Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Total Revenue")
plt.grid()
plt.show()
```

### 17.What is the median mobile price for each warranty period (Warranty_Period)?

```python
# Group by 'Warranty_Years' and calculate the median 'Price'
median_price_by_warranty = df.groupby('Warranty_Years')['Price'].median()
print("Median Mobile Price for Each Warranty Period:")
print(median_price_by_warranty)
```

### PIECHART

```python
# Group by 'Mobile_Brand' and sum up 'Quantity'
brand_sales = df.groupby('Mobile_Brand')['Quantity'].sum()
plt.figure(figsize=(8,8))
plt.pie(brand_sales, labels=brand_sales.index, autopct='%1.1f%%', startangle=140)
plt.title("Proportion of Mobile Brands Sold")
plt.show()
```
